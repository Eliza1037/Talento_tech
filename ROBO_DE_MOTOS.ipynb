```python
#Convertir el array sklearn a dataframe

df_iris=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target']=iris.target
```

```python
df_iris.columns
```

```python
df_iris.rename({'target':'species'}, inplace= True, axis=1) #inplace = True- guarda los cambios en el archivo original
df_iris
```

```python
#Transformo el tipo de dato del campo species

df_iris['species']=df_iris['species'].astype('category') #Covierte el tipo de dato de la columna species en "Category"
df_iris.info()
```

**Imputación** **de datos multivariados**

```python
#Asigno valores nulos a la columna de species


iris_amp=mf.ampute_data(df_iris, perc=0.25, random_state=100) #realizo una amputación de datos, haciendo que los nulos sean NaN, PERC=0.25-qUE EL 25% DE LOS DATOS SEAN VALORES NULOS- ASIGNO VALORES NULOS ALEATORIAMENTE.- RANDOM STATE= EL NUMERO QUE PONGO-LA POSIBILIDAD DEL ARBOL
iris_amp
```

```python
#CREAR KERNEL- ALGORITMO DONDE COGE COLUMNA POR COLUMNA E IDENTIFICA LOS VALORES y pone datos en donde hay NaN.
kds=mf.ImputationKernel(
    iris_amp,             #nombre de dataframe para entrenar modelo
    random_state=1991     #Valores aleatorios para entrenar
)

#Correr el algoritmo MICE durante 2 interacciones

kds.mice(2)  # en el parentesis va el numero de iteracciones

#Retorna el dataset completo

iris_completo= kds.complete_data()

```
